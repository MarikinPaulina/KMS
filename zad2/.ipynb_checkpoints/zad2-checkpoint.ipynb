```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.autonotebook import tqdm
import zad2
import pandas as pd
from matplotlib.animation import FuncAnimation
```

```python
N = 100
n = 1
kappa = 1
omega = np.pi**2 /2 *3
dt = 0.0001
s_data = int(1e2)
s_pos = int(1e2)
file_name = f'wyniki/z_polem/1_3'
steps = int(5e5)

sim = zad2.Simulation(n, N, kappa, omega, dt, s_data, s_pos, file_name, steps)
```

```python
sim.simulation()
```

```python
file_name = f'wyniki/z_polem/1_3'
data = pd.read_csv(file_name+'.out')

fig = plt.figure(figsize=(21,7))
ax = plt.subplot(131)
plt.plot(data['E'],'.')
ax.set_title("E")
ax = plt.subplot(132)
plt.plot(data['N'],'.')
ax.set_title("N")
ax = plt.subplot(133)
plt.plot(data['x'],'.')
ax.set_title("mean x")
```

```python
omega = []
E = []
for i in tqdm(range(-10,10)):
    omega.append(np.pi**2 * 0.5 *3 * (1+i/100))
    sim.omega = np.pi**2 * 0.5 *3 * (1+i/100)
    sim.file_name = f'wyniki/przejscie/{(1+i/100)}'
#     sim.simulation()
    
    data = pd.read_csv(sim.file_name+'.out')
    E.append(max(data['E']))
plt.figure(figsize=(7,7))
plt.plot(omega,E)
plt.ylabel('E')
plt.xlabel(r'$\omega$')
```
